```python
import pandas as pd
import logging
import joblib
import os
os.chdir('..')
from src.utils import load_data_to_predict
```

```python
MODEL_PATH='/Users/rohith/Documents/Projects/F1/model/model.joblib'

logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)
```

```python
def load_model():
    try:
        model = joblib.load(MODEL_PATH)
        logger.info("Model loaded successfully.")
        return model

    except Exception as e:
        raise Exception("Model could not be loaded. {e}")
```

```python
race = 20
year = 2025
```

```python
model = load_model()
data,driver = load_data_to_predict(race, year)
```

```python
predictions = model.predict(data)
```

```python
import datetime

def convert(n):
    try:
        n = int(n)
    except:
        n = float(n)
    return str(datetime.timedelta(seconds = n))
    
# Driver program
n = 76.49183291898339
print(convert(n))
```

```python
sorted_driver_preds = dict(sorted(zip(driver, predictions), key=lambda x: x[1]))
sorted_driver_preds = {k: convert(v) for k, v in sorted_driver_preds.items()}
for driver, pred in sorted_driver_preds.items():
    print(f"{driver}: {pred}")
```
