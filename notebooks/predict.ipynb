```python
import pandas as pd
import logging
import joblib
from src.utils import load_data_to_predict
```

```python
MODEL_PATH='./model/model.joblib'

logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)
```

```python
def load_model():
    try:
        model = joblib.load(MODEL_PATH)
        logger.info("Model loaded successfully.")
        return model

    except Exception as e:
        raise Exception("Model could not be loaded. {e}")
```

```python
race = 18
year = 2025
```

```python
model = load_model()
data,driver = load_data_to_predict(race, year)
```

```python
data
```

```python
predictions = model.predict(data)
```

```python
pred_int = predictions.round().astype(int)
sorted_driver_preds = dict(sorted(zip(driver, pred_int), key=lambda x: x[1]))
sorted_driver_preds
```
